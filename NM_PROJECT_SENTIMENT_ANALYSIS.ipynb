```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import seaborn as sns
from textblob import TextBlob

# Fixed: Using a predefined sentence instead of interactive input to prevent KeyboardInterrupt errors.
sentence = "i am so busy."

# Sentiment analysis function
def analyze_sentiment(text):
    sentiment = TextBlob(text).sentiment.polarity
    if sentiment > 0:
        return "Positive"
    elif sentiment < 0:
        return "Negative"
    else:
        return "Neutral"

# Perform sentiment analysis
sentiment_result = analyze_sentiment(sentence)

# Store sentiment data
sentiments = {
    "Positive": {"count": 0},
    "Negative": {"count": 0},
    "Neutral": {"count": 0}
}
sentiments[sentiment_result]["count"] += 1

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie([s["count"] for s in sentiments.values()], labels=sentiments.keys(), autopct='%1.1f%%', colors=['green', 'red', 'blue'])
plt.title("Sentiment Distribution - Pie Chart")
plt.show()

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(sentiments.keys(), [s["count"] for s in sentiments.values()], color=['green', 'red', 'blue'], edgecolor="black")
plt.title("Sentiment Analysis - Bar Chart")
plt.xlabel("Sentiment Type")
plt.ylabel("Count")
plt.show()

# Histogram
plt.figure(figsize=(8, 5))
plt.hist([s["count"] for s in sentiments.values()], bins=3, color='purple', edgecolor="black")
plt.title("Sentiment Analysis - Histogram")
plt.xlabel("Sentiment Count Range")
plt.ylabel("Frequency")
plt.show()

# Box Plot
plt.figure(figsize=(8, 5))
plt.boxplot([s["count"] for s in sentiments.values()], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Sentiment Count Distribution - Box Plot")
plt.xlabel("Sentiment Count")
plt.show()

# Heatmap
sentiment_df = pd.DataFrame([s["count"] for s in sentiments.values()], index=sentiments.keys(), columns=["Count"])
plt.figure(figsize=(6, 4))
sns.heatmap(sentiment_df, annot=True, cmap="coolwarm", linewidths=0.5, fmt='d')
plt.title("Sentiment Analysis - Heatmap")
plt.show()

# Table Representation
df = pd.DataFrame(list(sentiments.items()), columns=["Sentiment", "Data"])
df["Count"] = df["Data"].apply(lambda x: x["count"])
df.drop(columns=["Data"], inplace=True)
print("\n🔹 Sentiment Analysis Table 🔹")
print(df.to_string(index=False))

# Tree Representation
G = nx.Graph()
G.add_edge("Sentiment Analysis", sentiment_result)
plt.figure(figsize=(8, 5))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', font_size=10)
plt.title("Sentiment Tree Structure")
plt.show()

# Emoji-Based Output
emoji_map = {"Positive": "😊", "Negative": "😡", "Neutral": "😐"}
print(f"\n🔹 Sentiment in Emojis: {emoji_map.get(sentiment_result, '❓')} {sentiment_result}")

# Final Output
print(f"\n🔹 Analyzed Sentence: \"{sentence}\"")
print(f"🔹 Detected Sentiment: {sentiment_result}")
```
